# Lists and function
- a list is a single object, so when it is passed in, it only requires a single parameter value

```python
def modify_elements(items):
    '''Multyply all elements in a list by 2'''
    for i in range(len(items)):
        items[i] *= 2
        # or 
        #items[i] = item[i] * 2
        return items
        
        
numbers = [10,3,7,1,9]
modify_elements(numbers)

print(numbers)
```

### List methods
##### Try these list methods

- a_list.sort()
- a_list.index(value)
- a_list.insert(index,value)
- a_list.append(value)
- a_list.remove(value)
- a_list.clear()
- a_list.count(value)
- a_list.reverse()

```python
# .sort() modifies the original list without explicit reassignment
a_list = [5,6,4,3]
a_list.sort()
print(a_list)      
```

```python
# here's another way to do it
a_list = [5,6,4,3]
sorted_list = sorted(a_list)
print(sorted_list)  
print(a_list)       
```

```python
# searching sequences by the number you are searching for with .index(value)
a = [-47, 76, 80, 50, 72, 111, -10]
a.index(50) 
```

```python
# .insert(index, value)
a = [-47, 76, 80, 50, 72, 111, -10]
a.insert(3,100)
print(a)
```

```python
# .append('value')
a = [-47, 76, 80, 50, 72, 111, -10]
a.append(100)
print(a)
```
