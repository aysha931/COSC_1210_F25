### Stripping white space

Later we will learn how to process text files with Python. To do so, we can strip the white spaces so we can process the string. Those white spaces include
- ' ' or '\s'- spaces
- '\n' - new lines
- '\t' - tabs

```python
astring = "     Python is a versatile, readable language  .\t... \n"
# .strip() takes out all white spaces from the ends
print("original")
print(astring)
print("*")
print("stripped")
print(astring.strip())
print("*")
```

```python
astring = "     Python is a versatile, readable language  .\t... \n"
print(astring.lstrip())
print("*")
```

```python
astring = "     Python is a versatile, readable language  .\t... \n"
print(astring.rstrip())
print("*")
```

```python
def main():
    '''
    Input - a text file with first names and numbers separated by a comma
    Process - read in the file, parse the file, add a last name as input from the user
    add that to the parsed information
    Output - write the first, last, and number to a new file
    '''
    input_filename = "names_numbers.txt"
    output_filename = "updated_names_numbers.txt"
    updated_data = []

    try:
        with open(input_filename, 'r') as infile:
            print(f"Successfully opened {input_filename} for reading.")
            for line in infile:
                parts = line.strip().split('')
                if len(parts) ==2:
                    first_name, number = parts[0], parts[1]
                    last_name = input(f"Enter a last name for {first_name}:")
                    updated_data.append(f"{first_name} {last_name}")
                else:
                    print(f"Warnin: Skipping malformed line in {input_filename}: {line.strip()}")
    except FileNotFoundError:
        print(f"Error: The file ' {input_filename}' was not found. Please make sure it's in the same directory as the program.")
        print("please create 'names_numbers.txt' with content like: 'Alice 123' then run again")
        return 
    except Exception as e:
        print(f" an unxpected error occured while reading the file: {e}")
        return
    try:
        with open(output_filename, 'w') as outfile:
            for entry in updated_data:
                outfile.write(entry + '\n')
        print(f"Successfully wrote updated information to '{output_filename}'.")
    except Exception as e:
        print(f"an error occurred while writing to the output file: {e}")
    main()
```

```python
class LibraryBook:
    def __init__(self, book_title, borrower_name, weeks_until_due):
        self.book_title = book_title
        self.borrower_name = borrower_name
        self.weeks_until_due = weeks_until_due

    def renew(self, num_weeks):
        self.weeks_until_due += num_weeks
        print(f"'{self.book_title}' renewed for {num_weeks} weeks. New due date: {self.weeks_until_due} weeks.")

    def return_book(self):
        self.borrower_name = ""
        self.weeks_until_due = ""
    def display_info(self):
        print(self.book_title, self.borrower_name, self.weeks_until_due)

def main():
    book1 = LibraryBook("The Great Gatsby", "Alex", 3)
    book2 = LibraryBook("Frankenstein", "jordan", 0)

    book1.renew(5)
    book2.return_book()

    print("The current library records are:")
    book1.display_info()
    book2.display_info()


main()
```
