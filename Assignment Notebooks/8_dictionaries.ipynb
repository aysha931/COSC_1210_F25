### Basic dictionary statements and methods

```python
## removing key-value pairs

# del - simply remove a key-value pair by calling on its key
phone_dict = {"fatima": 4378099454, "bob": 16129994444, "marlo": 549823460}

# note the square brackets for accessing a value related to a key
del phone_dict["bob"]
print(phone_dict)
```

```python
# pop - return the value from the removed key
phone_dict = {"fatima": 4378099454, "bob": 16129994444, "marlo": 549823460}

# note the round brackets used for methods
removed_item = phone_dict.pop("bob")
print("we removed", removed_item)
print("the new dictionary is", phone_dict)
```

### Sets

Sets are unordered collections of unique values. They are denoted with curly brackets
- Notice that after the set expression is evaluated, they order is changed...

```python
colors = {'red', 'red', 'orange', 'yellow', 'blue', 'red', 'green', 'yellow'}
print(colors)
```

```python
# create a set with the set() function
numbers = list(range(10)) + list(range(5))
print(numbers)
numbers_set = set(numbers)
print(numbers_set)
```

```python
# union - all unique elements from two sets
set1 = {1, 2, 4, 5, 6, 7, 11}
set2 = {1, 3, 4, 7, 8, 11}

set1 | set2
```

```python
# union
set1 = {1, 2, 4, 5, 6, 7, 11}
set2 = {1, 3, 4, 7, 8, 11}

set1.union(set2)
```

```python
# intersection - unique elements sets have in common

set1 = {1, 2, 4, 5, 6, 7, 11}
set2 = {1, 3, 4, 7, 8, 11}

set1 & set2
```

```python
# intersection - unique elements sets have in common

set1 = {1, 2, 4, 5, 6, 7, 11}
set2 = {1, 3, 4, 7, 8, 11}

set1.intersection(set2)
```

```python
# difference - unique elements in the left set and not in the right set

set1 = {1, 2, 4, 5, 6, 7, 11}
set2 = {1, 3, 4, 7, 8, 11}

set1 - set2
```

```python
# difference - unique elements in the left set and not in the right set

set1 = {1, 2, 4, 5, 6, 7, 11}
set2 = {1, 3, 4, 7, 8, 11}

set1.difference(set2)
```

```python
# other set function
set1 = {1, 2, 4, 5, 6, 7, 11}
set1.add(17)
set1.add(3)
print(set1)
```

```python
## Try the following set methods
set1 = {1, 2, 4, 5, 6, 7, 11}
# .remove()
set1.remove(1)
print(set1)

# .pop()
set1.pop()
print(set1)

# .clear()
set1.clear()
print(set1)
```
